```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=20, random_state=10)
```

```python
plt.scatter(X, y)
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color='red')
```

```python
# Iteration 0
# Lets apply Gradient Descent assuming slope is constant m = 16.14529159
# and let's assume the starting value for intercept b = 100
y_pred = ((16.145 * X) + 100).reshape(4)
```

```python
y_pred.shape
```

```python

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
     
```

```python
# Iteration 1
m = 16.145
b = 50
learning_rate = 0.1

slope_b = np.sum(m * X.ravel() + b - y) * (1/X.shape[0])
slope_b
```

```python
b = b - slope_b * learning_rate
b
```

```python
y_pred1 = (16.145 * X + b).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a65a', label='b=100')
plt.plot(X, y_pred1, color='#f39c12', label='b={}'.format(round(b,2)))
plt.legend()
plt.show()
```

```python
#Iteration 2
slope_b = np.sum(m * X.ravel() + b - y) * (1/X.shape[0])
slope_b
```

```python
b = b - slope_b * learning_rate
b
```

```python
y_pred2 = (16.145 * X + b).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a65a', label='b=100')
plt.plot(X, y_pred1, color='#f39c12', label='b={}'.format(round(b,2)))
plt.plot(X, y_pred2, color='#f39c12', label='b={}'.format(round(b,2)))
plt.legend()
plt.show()
```

```python
#Iteration 3
slope_b = np.sum(m * X.ravel() + b - y) * (1/X.shape[0])
slope_b
```

```python
b = b - slope_b * learning_rate
b
```

```python
y_pred3 = (16.145 * X + b).reshape(4)
```

```python
plt.figure(figsize=(15,15))
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#00a65a', label='b=100')
plt.plot(X, y_pred1, color='#f39c12', label='b={}'.format(round(b,2)))
plt.plot(X, y_pred2, color="#12bbf3", label='b={}'.format(round(b,2)))
plt.plot(X, y_pred3, color="#bbf312", label='b={}'.format(round(b,2)))
plt.legend()
plt.show()
```

```python
b = 100
m = 16.145
lr = 0.1

epochs = 100

for i in range(epochs):
  slope_b = np.sum(m * X.ravel() + b - y) * (1/X.shape[0])
  b = b - (lr * slope_b)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)
```
